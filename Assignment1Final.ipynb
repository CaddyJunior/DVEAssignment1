```python
#1. Identify what percentage of gallons, miles, and odometer entries are missing.
df_new_updated = filtered_df.copy()
# List of columns to check
cols_to_check = ['gallons', 'miles', 'odometer']

# Calculate missing percentages
missing_percentages = df_new_updated[cols_to_check].isnull().mean() * 100

# Display results
for col, pct in missing_percentages.items():
    print(f"{col}: {pct:.2f}% missing")
```

```python
#2. The miles, gallons and mpg columns are interdependent. If one is missing, the other two can be used to calculate it.
#Clean and convert numeric fields
def clean_and_convert(val):
    if isinstance(val, str):
        val = val.replace(',', '')  # Remove thousands separator
    try:
        return float(val)
    except (ValueError, TypeError):
        return np.nan  # Return NaN for invalid entries

# Apply cleaning to relevant columns
for col in ['miles', 'gallons', 'mpg', 'odometer']:
    df_new_updated[col] = df_new_updated[col].apply(clean_and_convert)

#Fill missing values using interdependent logic
def fill_interdependent(row):
    miles, gallons, mpg = row['miles'], row['gallons'], row['mpg']
    
    # Fill mpg
    if pd.isna(mpg) and pd.notna(miles) and pd.notna(gallons) and gallons != 0:
        row['mpg'] = miles / gallons
    
    # Fill gallons
    elif pd.isna(gallons) and pd.notna(miles) and pd.notna(mpg) and mpg != 0:
        row['gallons'] = miles / mpg
    
    # Fill miles
    elif pd.isna(miles) and pd.notna(gallons) and pd.notna(mpg):
        row['miles'] = mpg * gallons
    
    return row

# Apply row-wise imputation
df_new_updated = df_new_updated.apply(fill_interdependent, axis=1)
```

### 2 Feature Engineering
- We can use the existing features to create new features with more useable information. Add the following features:
1. Create a new column with the currency. (Something to keep in mind is
that the Swiss Franc has a period in the abbreviation).
2. Create a new column containing the float value of the total spend and the
cost per gallon. (Swiss Franc comment as above).
3. Car make, model, year, User ID: use the url (the last value in the URL is
the user ID)
- The data is given in imperial units, and in SA, we use proper measurement
standards.
1. litres filled: use the gallons - consider whether to use UK or US gallons.
2. km driven: use the miles driven to compute this.
3. litres per 100km: use the two new features to calculate this.

```python
#1. Create a new column with the currency. (Something to keep in mind is that the Swiss Franc has a period in the abbreviation).
# Sample column: 'cost_per_gallon' or 'total_spent'
def extract_currency(val):
    if isinstance(val, str):
        match = re.match(r'([^\d\s]+)', val.strip())
        if match:
            return match.group(1)
    return None

# Apply to both cost fields
df_new_updated['currency'] = df_new_updated['cost_per_gallon'].apply(extract_currency)
```

```python
df_new_updated[['cost_per_gallon','currency']].head(10)
```

```python
#2. Create a new column containing the float value of the total spend and the cost per gallon. (Swiss Franc comment as above). 
#Function to extract numeric value from a string with currency
def extract_float(val):
    if isinstance(val, str):
        # Remove currency symbols/letters and commas
        cleaned = re.sub(r'[^\d.,]', '', val).replace(',', '')
        try:
            return float(cleaned)
        except ValueError:
            return np.nan
    return val if isinstance(val, (int, float)) else np.nan

# Apply to cost columns
df_new_updated['total_spent_float'] = df_new_updated['total_spent'].apply(extract_float)
df_new_updated['cost_per_gallon_float'] = df_new_updated['cost_per_gallon'].apply(extract_float)
```

```python
df_new_updated[['total_spent','total_spent_float', 'cost_per_gallon','cost_per_gallon_float']].head(10)
```

```python
#3. Car make, model, year, User ID: use the url (the last value in the URL is the user ID)
# Function to extract components from the URL
def extract_vehicle_info(url):
    if isinstance(url, str):
        parts = url.strip().split('/')
        if len(parts) >= 5:
            return pd.Series({
                'car_make': parts[-4],
                'car_model': parts[-3],
                'car_year': parts[-2],
                'user_id': parts[-1]
            })
    return pd.Series({'car_make': None, 'car_model': None, 'car_year': None, 'user_id': None})

# Apply to the user_url column
df_new_updated[['car_make', 'car_model', 'car_year', 'user_id']] = df_new_updated['user_url'].apply(extract_vehicle_info)
```

```python
df_new_updated[['user_url','car_make', 'car_model', 'car_year', 'user_id']].head(10)
```

```python
# litres filled: use the gallons - consider whether to use UK or US gallons.
# Conversion constants
US_GALLON_TO_LITRES = 3.78541
MILES_TO_KM = 1.60934

# Convert gallons to litres
df_new_updated['litres_filled'] = df_new_updated['gallons'].apply(lambda g: float(g) * US_GALLON_TO_LITRES if pd.notna(g) else np.nan)
df_new_updated['cost_per_liter_float'] = df_new_updated['cost_per_gallon_float'].apply(lambda g: float(g) * US_GALLON_TO_LITRES if pd.notna(g) else np.nan)
df_new_updated['total_spent_per_liter_float'] = df_new_updated['total_spent_float'].apply(lambda g: float(g) * US_GALLON_TO_LITRES if pd.notna(g) else np.nan)

# 2. km driven: use the miles driven to compute this. 
# Convert miles to kilometers
df_new_updated['km_driven'] = df_new_updated['miles'].apply(lambda m: float(m) * MILES_TO_KM if pd.notna(m) else np.nan)

#3. litres per 100km: use the two new features to calculate this.
#Calculate litres per 100km
def calc_l_per_100km(row):
    if pd.notna(row['litres_filled']) and pd.notna(row['km_driven']) and row['km_driven'] != 0:
        return (row['litres_filled'] / row['km_driven']) * 100
    return np.nan

df_new_updated['litres_per_100km'] = df_new_updated.apply(calc_l_per_100km, axis=1)
```

These are based on typical fuel prices and tank sizes in each region

| Currency | Min Cost/Litre | Litres Range | Max Spend |
|----------|------------------|--------------|-------------------|
| `$` (USD)   | `$`1.35     | 1–120        |  `$`162.16         |
| `£` (GBP)   | £1.06     | 1–120        | £127.66          |
| `€` (EUR)   | €1.25     | 1–120        | €150.00         |
| `CA$` (CAD) | `CA$`1.85 | 1–120        | `CA$`222.22      |
| `R` (ZAR)   | R25         | 1–120        | R3000       |


```python
def remove_outliers_by_currency(df):
    thresholds  = {
        '$': {
            'min_spend': round(25 / 18.5, 2),   # ≈ $1.35
            'max_spend': round(3000 / 18.5, 2), # ≈ $162.16
            'min_litres': 1,
            'max_litres': 120,
            'min_cpl': round(25 / 18.5, 2),     # ≈ $1.35
            'max_cpl': round(3000 / 18.5, 2)    # ≈ $162.16
        },
        '£': {
            'min_spend': round(25 / 23.5, 2),   # ≈ £1.06
            'max_spend': round(3000 / 23.5, 2), # ≈ £127.66
            'min_litres': 1,
            'max_litres': 120,
            'min_cpl': round(25 / 23.5, 2),     # ≈ £1.06
            'max_cpl': round(3000 / 23.5, 2)    # ≈ £127.66
        },
        '€': {
            'min_spend': round(25 / 20.0, 2),   # ≈ €1.25
            'max_spend': round(3000 / 20.0, 2), # ≈ €150.00
            'min_litres': 1,
            'max_litres': 120,
            'min_cpl': round(25 / 20.0, 2),     # ≈ €1.25
            'max_cpl': round(3000 / 20.0, 2)    # ≈ €150.00
        },
        'CA$': {
            'min_spend': round(25 / 13.5, 2),   # ≈ CA$1.85
            'max_spend': round(3000 / 13.5, 2), # ≈ CA$222.22
            'min_litres': 1,
            'max_litres': 120,
            'min_cpl': round(25 / 13.5, 2),     # ≈ CA$1.85
            'max_cpl': round(3000 / 13.5, 2)    # ≈ CA$222.22
        },
        'R': {
            'min_spend': 25,
            'max_spend': 3000,
            'min_litres': 1,
            'max_litres': 120,
            'min_cpl': 25,
            'max_cpl': 3000
        }
    }

    cleaned_frames = []

    for currency, limits in thresholds.items():
        subset = df[df['currency'] == currency]

        filtered = subset[
            (subset['total_spent_per_liter_float'] >= limits['min_spend']) & (subset['total_spent_per_liter_float'] <= limits['max_spend']) &
            (subset['litres_filled'] >= limits['min_litres']) & (subset['litres_filled'] <= limits['max_litres']) &
            (subset['cost_per_liter_float'] >= limits['min_cpl']) & (subset['cost_per_liter_float'] <= limits['max_cpl'])
        ]

        cleaned_frames.append(filtered)

    return pd.concat(cleaned_frames, ignore_index=True)

cleaned_data = remove_outliers_by_currency(df_new_updated)
```

```python
#4. Now reduce your dataset to only the entries on the 1st Tuesday and 1st Wednesday in SA every month.

sa_data['month'] = sa_data['timestamp'].dt.month
sa_data['year'] = sa_data['timestamp'].dt.year

#Get fuel price per day
daily_prices = sa_data.groupby('date')['cost_per_liter_float'].mean().reset_index()

# Identify 1st Tuesday and 1st Wednesday of each month
def get_first_weekdays(year, month):
    dates = pd.date_range(start=f'{year}-{month:02d}-01', end=f'{year}-{month:02d}-07')
    first_tuesday = dates[dates.dayofweek == 1].min()  # Tuesday = 1
    first_wednesday = dates[dates.dayofweek == 2].min()  # Wednesday = 2
    return first_tuesday.date(), first_wednesday.date()

first_days = []
for year in sa_data['year'].unique():
    for month in sa_data[sa_data['year'] == year]['month'].unique():
        try:
            tuesday, wednesday = get_first_weekdays(year, month)
            first_days.append({'year': year, 'month': month, 'tuesday': tuesday, 'wednesday': wednesday})
        except:
            continue

```

```python
# 5. For each Tuesday and Wednesday, add an indicator for whether the price
# goes up or the price goes down that month

first_days_df = pd.DataFrame(first_days)

# Merge price info and compute price change
first_days_df = first_days_df.merge(daily_prices.rename(columns={'date': 'tuesday', 'cost_per_liter_float': 'price_tuesday'}), on='tuesday', how='left')
first_days_df = first_days_df.merge(daily_prices.rename(columns={'date': 'wednesday', 'cost_per_liter_float': 'price_wednesday'}), on='wednesday', how='left')
first_days_df['price_change'] = np.where(first_days_df['price_wednesday'] > first_days_df['price_tuesday'], 'up', 'down')

first_days_df.head(10)
```

```python
#Filter SA data to only 1st Tuesday and Wednesday entries
target_dates = pd.concat([
    first_days_df[['tuesday', 'price_change']].rename(columns={'tuesday': 'date'}),
    first_days_df[['wednesday', 'price_change']].rename(columns={'wednesday': 'date'})
])
sa_data_filtered = sa_data.merge(target_dates, on='date', how='inner')
sa_data_filtered.head(10)
```

```python
# 6. Do more people refuel on the first Wednesday of the month when the prices goes down? Compare refueling counts on Wednesdays when price goes down
wednesday_down = sa_data_filtered[(sa_data_filtered['weekday'] == 'Wednesday') & (sa_data_filtered['price_change'] == 'down')]
wednesday_up = sa_data_filtered[(sa_data_filtered['weekday'] == 'Wednesday') & (sa_data_filtered['price_change'] == 'up')]

print(f"Refueling events on 1st Wednesday when price goes DOWN: {len(wednesday_down)}")
print(f"Refueling events on 1st Wednesday when price goes UP: {len(wednesday_up)}")
```

```python
# Plot comparison
import matplotlib.pyplot as plt

plt.bar(['Price Down', 'Price Up'], [len(wednesday_down), len(wednesday_up)], color=['green', 'red'])
plt.title('Refueling on 1st Wednesday by Price Change Direction')
plt.ylabel('Number of Refueling Events')
plt.tight_layout()
plt.show()
```



Yes, slightly more people refuel on the first Wednesday of the month when the fuel price goes down.

The green bar labeled "Price Down" is visibly higher than the red "Price Up" bar, indicating a modest increase in refueling activity when prices drop. This supports the hypothesis that South African drivers respond to fuel price reductions by postponing refueling until Wednesday, likely to take advantage of the lower cost. This behavior reflects price sensitivity and consumer awareness of fuel price adjustments. While the difference isn’t dramatic, it’s statistically and behaviorally meaningful—especially for modeling user decision-making or designing fuel-related incentives.
